# Data Cleaning Example
This notebook demonstrates basic data loading, cleaning, and visualization.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', None, 'Eve'],
    'age': [25, None, 35, 28, None],
    'score': [88, 92, None, 70, 95]
})
df
```

## Clean missing values

```
df_clean = df.dropna()
df_clean
```

## Simple Visualization

```
plt.figure(figsize=(6,4))
plt.plot(df_clean['age'], df_clean['score'])
plt.xlabel('Age')
plt.ylabel('Score')
plt.title('Score vs Age')
plt.show()
```
